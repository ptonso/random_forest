# Extracting rules from Random Forest

```python
# Importing

import pandas as pd

from sklearn.ensemble import RandomForestClassifier

from pkg import utils
from pkg import explain
```

```python
# Importing model

parent = 'models'
filename = 'v01.pkl'

path = utils.get_path(parent, filename)

rf = utils.read_model(path)
features = rf.feature_names_in_
```

```python
# Importing feature df

parent = 'data'
filename = 'featured.csv'

path = utils.get_path(parent, filename)

df = pd.read_csv(path, index_col = 0)

y = df.iloc[:,0]
X = df.iloc[:,1:]
```

```python
# Configurations

bins = 20
```

```python
feature = X.iloc[:,0]
feature.quantile(0.2)
```

```python
# Creating binning df

bin_df = explain.binning_df(X, bins=bins)

print(bin_df.to_markdown())
```

```python
rules = []

for tree in rf:
    rules += explain.find_rules(tree, features)

print('total number of rules: ', len(rules))

unique_rules = []
for rule in rules:
    if rule not in unique_rules:
        unique_rules.append(rule)

print('total number of unique rules: ', len(unique_rules))

print('repeated rules: ', len(rules) - len(unique_rules))

for i in range(10):
    print(rules[i])
```
