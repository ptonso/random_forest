# Featuring

**Import** data from "data/cleaned.csv"

**Export** to "data/featured.csv"

```python
# Setup

import numpy as np
import pandas as pd

from pkg import utils
from pkg import feature
```

```python
# Importing data

path = utils.get_parents() + r"/data/cleaned.csv"

df = pd.read_csv(path, index_col=0)
df.head()
```

```python
X_list = df.columns[1:-5]
y_list = df.columns[-1]

X = df[X_list]
y = df[y_list]
```

```python
featured = pd.concat([y,X],axis=1)
featured.head()
```

```python
# Exporting data

featured = featured.dropna(axis=0)

run_save = False
if not run_save:
  utils.save_df(featured, "featured.csv")
  run_save = True
```
