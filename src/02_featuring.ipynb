# Featuring

**Import** data from "data/cleaned.csv"

**Export** to "data/featured.csv"

```python
# Setup

import numpy as np
import pandas as pd

from pkg import utils
from pkg import feature
```

```python
# Importing data

parent = 'data'
filename = 'cleaned.csv'

path = utils.get_path(parent, filename)

df = pd.read_csv(path, index_col=0)
df.head()
```

```python
# Configurations

quantile = 0.2
less_than = True
```

```python
X = df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
y = feature.threshold_mapping(y, quantile=quantile, less_than=less_than)

print(y)
```

```python
featured = pd.concat([y,X],axis=1)
featured.head()
```

```python
# Exporting data

featured = featured.dropna(axis=0)

run_save = False
if not run_save:
  utils.save_df(featured, "featured.csv")
  run_save = True
```
