# Visualizing dataset using PCA dimensional reduction

```python
# Setup

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d
from sklearn import decomposition
```

```python
parent_abs = os.path.abspath(os.path.join(os.getcwd(),os.pardir))
df = pd.read_csv(os.path.join(parent_abs,'data\\featured.csv'),index_col=0)
```

```python
y = df.iloc[:,0]
```

```python
pca = decomposition.PCA(n_components=2)
view = pca.fit_transform(df)
view.shape
```

```python
# 2d Plot (What the fuck does it means?)

plt.scatter(view[:,0], view[:,1],c=y)
plt.xlabel('PCA-1')
plt.ylabel('PCA-2')
```
