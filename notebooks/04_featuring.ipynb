# Creating features for Model
This notebook takes 'cleaned.csv'

and return 'featured.csv'

```python
# Setup

import os

import numpy as np
import pandas as pd
```

```python
parent_abs = os.path.abspath(os.path.join(os.getcwd(),os.pardir))
df = pd.read_csv(os.path.join(parent_abs,'data\\cleaned.csv'),index_col=0)
```

```python
X_list = df.columns[1:-5]
y_list = df.columns[-1]

X = df[X_list]
y = df[y_list]
```

```python
featured = pd.concat([y,X],axis=1)
featured.head()
```

```python
relative_path = r"data\featured.csv"

path = os.path.join(parent_abs,relative_path)

print(f"\n featured path {path}")

featured.to_csv(path)
```
